```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df14 = pd.read_excel(r"D:\Data\esc\ESC-2014-grand_final-full_results.xls", skiprows=1, na_values=['\n','-']).fillna(0)
df14['Year'] = 2014
df15 = pd.read_excel(r"D:\Data\esc\ESC-2015-grand_final-full_results.xls", skiprows=1, na_values=['\n','-']).fillna(0)
df15['Year'] = 2015
df16 = pd.read_excel(r"D:\Data\esc\ESC-2016-grand_final-full_results.xls", skiprows=1, na_values=['\n','-']).fillna(0)
df16['Year'] = 2016
df16['Points'] = df16['Jury Points'] + df16['Televote Points']

df = pd.concat((df14, df15, df16), 0)
df.head()
```

```python
df['jury_disagreement'] = (df[[c for c in df.columns if c.startswith('Jury')]]/df[[c for c in df.columns if c.startswith('Jury')]].mean()).std(1)
df.head()
```

```python
dfp = pd.pivot_table(df, index=['Year', 'To country'], columns=['From country'], values='Points').fillna(0)
dfp['tot_points'] = dfp.sum(1)
dfp.tail()
```

```python
countries = df['From country'].unique()

df_co_cor = pd.DataFrame(index=countries, columns=['cor%d' %y for y in df['Year'].unique()] + ['cor'])
for country in countries:
    for y in range(2014, 2017):
        df_co_cor.loc[country, 'cor%d' %y] = np.corrcoef(dfp.loc[(y, ), country].fillna(0), dfp.loc[(y, ), 'tot_points'])[0, 1]
    df_co_cor.loc[country, 'cor'] = df_co_cor.loc[country, 'cor'] = np.corrcoef(dfp.loc[:, country].fillna(0), dfp.loc[:, 'tot_points'])[0, 1]
    
df_co_cor.sort('cor', ascending=False)
```
